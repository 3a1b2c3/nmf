```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
plt.axes().set_aspect('equal', 'box')
plt.plot(x * y, y * y)
plt.show()
```

```python
l = np.sqrt(x**2 + y**2)
# plt.plot(x / l, y / l)
print(l)
plt.plot(x / l, y / l)
plt.show()
```
